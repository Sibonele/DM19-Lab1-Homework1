```python
#lower string
dataset=data_file['review'].str.lower()

# Preparing the text column in data_file for bag of words, a corpus is created first
corpus = []
for words in dataset:
     corpus.append(words)
print(corpus)

vectorizer = CountVectorizer()

words=vectorizer.fit_transform(corpus).todense(10)
counted_words= vectorizer.vocabulary_
print(words) 
print(counted_words)
```

```python
#creating a dictionary of the counted words
word2count = {} 
for data in counted_words: 
    words = nltk.word_tokenize(data) 
    for word in words: 
        if word not in word2count.keys(): 
            word2count[word] = 1
        else: 
            word2count[word] += 1
print(word2count)
```

```python
#Retrieving the first 1000 words which occure the most
import heapq 
freq_words = heapq.nlargest(1000, word2count, key=word2count.get)
print(freq_words)
```
